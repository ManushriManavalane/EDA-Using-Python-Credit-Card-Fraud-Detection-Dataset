```python
 
#Data Visualization
def draw_histograms(dataframe, features, rows, cols):
    fig = plt.figure(figsize=(20, 20))
    for i, feature in enumerate(features):
        ax = fig.add_subplot(rows, cols, i+1)
        dataframe[feature].hist(bins=20, ax=ax, facecolor='midnightblue')
        ax.set_title(feature + " Distribution", color="DarkRed")
        ax.set_yscale("log")
    fig.tight_layout()
    plt.show()

draw_histograms(data, data.columns, 8, 4)

```

```python
import matplotlib.pyplot as plt

def draw_boxplots(dataframe, features, rows, cols):
    fig = plt.figure(figsize=(20, 20))
    for i, feature in enumerate(features):
        ax = fig.add_subplot(rows, cols, i+1)
        dataframe.boxplot(column=feature, ax=ax, vert=False)
        ax.set_title(feature + " Distribution", color="DarkRed")
    fig.tight_layout()
    plt.show()

draw_boxplots(data, data.columns, 8, 4)
```

```python
#Feature Analysis
import matplotlib.pyplot as plt
import seaborn as sns

def plot_boxplots(dataframe, feature, target):
    plt.figure(figsize=(8, 6))
    sns.boxplot(x=target, y=feature, data=dataframe)
    plt.title(f"{feature} Distribution - Non-Fraud vs Fraud", fontsize=16)
    plt.xlabel("Fraudulent Transaction", fontsize=12)
    plt.ylabel(feature, fontsize=12)
    plt.show()

# Example usage:
plot_boxplots(data, "Amount", "Class")
```

```python
#Class Imbalance Analysis
import pandas as pd

def class_imbalance_analysis(dataframe, target):
    class_counts = dataframe[target].value_counts()
    positive_count = class_counts[1]
    negative_count = class_counts[0]
    imbalance_ratio = positive_count / negative_count

    print("Class Imbalance Analysis:")
    print("=========================")
    print("Positive (Fraudulent) Instances:", positive_count)
    print("Negative (Non-Fraudulent) Instances:", negative_count)
    print("Imbalance Ratio (Positive to Negative):", imbalance_ratio)

# Example usage:
class_imbalance_analysis(data, "Class")
```

```python
#Outliers
import pandas as pd

def data_quality_check(dataframe):
   
    # Identify outliers (assuming numerical columns)
    numerical_columns = dataframe.select_dtypes(include='number').columns
    outliers = []
    for column in numerical_columns:
        q1 = dataframe[column].quantile(0.25)
        q3 = dataframe[column].quantile(0.75)
        iqr = q3 - q1
        lower_bound = q1 - (1.5 * iqr)
        upper_bound = q3 + (1.5 * iqr)
        column_outliers = dataframe[(dataframe[column] < lower_bound) | (dataframe[column] > upper_bound)]
        outliers.append(column_outliers)
    print("Outliers:")
    print("==========")
    print(outliers)
 
data_quality_check(data)
```
